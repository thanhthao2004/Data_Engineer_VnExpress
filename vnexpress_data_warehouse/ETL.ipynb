### Lấy thứ, giờ tại cột date

```python
# Define the regex pattern for valid rows
pattern = r'^(Thứ \w+|Chủ nhật), \d{1,2}/\d{1,2}/\d{4}, \d{1,2}:\d{2} \(GMT\+7\)$'
df['date'] = df['date'].astype(str)

# Find rows that do not match the pattern
invalid_rows = df[~df['date'].str.match(pattern)]

# Display invalid rows
print(invalid_rows['date'])
```

```python
import pandas as pd
from datetime import datetime
# Function to extract date, day of week, and time
def extract_date_info(row):
    # Split the string by ', ' to get the components
    print(row)
    split_data = row.split(', ')
    
    # Extract the date using the second part (format: '8/6/2024')
    date = datetime.strptime(split_data[1], '%d/%m/%Y')
    
    # Extract the day of the week from the first part (e.g., 'Thứ bảy')
    date_in_week = split_data[0]
    
    # Extract the time from the third part (e.g., '12:00')
    time = split_data[2].split(' ')[0]
    
    return pd.Series([date, date_in_week, time])

# Apply the function to the dataframe and create new columns
df[['date', 'date_in_week', 'time']] = df['date'].apply(extract_date_info)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Khởi tạo TfidfVectorizer
vectorizer = TfidfVectorizer()

# Chuyển đổi dữ liệu thành ma trận TF-IDF
X = vectorizer.fit_transform(df['content'])

# Lấy từ khóa và điểm TF-IDF
features = vectorizer.get_feature_names_out()
scores = X.sum(axis=0).A1

# Tạo DataFrame từ từ khóa và điểm TF-IDF
keywords_df = pd.DataFrame({'keyword': features, 'score': max(scores)})

# Sắp xếp theo điểm TF-IDF giảm dần
keywords_df = keywords_df.sort_values(by='score', ascending=False)

print("Các từ khóa trọng tâm:")
print(keywords_df)  
```

```python
import random
def random_name(df):

  # Tạo danh sách các tên duy nhất
  unique_names = df['author'].dropna().unique().tolist()

  # Hàm để chọn một tên ngẫu nhiên
  def get_random_name():
    return random.choice(unique_names)

  # Áp dụng hàm lên các giá trị NaN
  df['author'] = df['author'].fillna(df['author'].apply(lambda x: get_random_name() if pd.isna(x) else x))

  return df

df = random_name(df)
```
